```python
# DependÃªncias
from components import Network, Controller
from components.qkd.app import *
import numpy as np
```

```python
# Definindo rede e controlador
rede = Network()
rede.set_topology("China")
controlador = Controller(rede)

# Algumas propriedades da rede
rede.neprs = 3
rede.set_fidelity(0.95)
rede.draw()
```

```python
# Propriedades da simulaÃ§Ã£o
apps = ["BB84", "E91", "B92"]
apps_distribution = [0.33, 0.33, 0.33]
routes_calculation_type = 'kshortest'
numero_qubits = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
requests = 100
n_simulacoes = 10
caso = 5
```

```python
# Executando simulaÃ§Ãµes variando o nÃºmero de qubits com k = 1
k = 1
for nqubits in numero_qubits:
    rede.set_nqubits(nqubits)
    vazao, taxa = run_simulations(rede, controlador, n_simulacoes, requests, apps, apps_distribution, caso, routes_calculation_type, k)
    media_vazao = np.mean(vazao)
    desvio_padrao_vazao = np.std(vazao)
    print(f'=SPLIT("{formatar_numero(media_vazao)} {formatar_numero(desvio_padrao_vazao)}"; " ")')
```
